```python
def group_category(category): 
    if category in ['Office Products', 'Electronics', 'Tools & Home Improvement', 'Cell Phones & Accessories', 'Arts, Crafts & Sewing']: 
        return category
    else: 
        return 'Others'
amazon_data["root_bs_category"] = amazon_data["root_bs_category"].apply(lambda category: group_category(category))
```

```python
dict_category={'Office Products':1, 'Electronics':2, 'Tools & Home Improvement':3,
       'Cell Phones & Accessories':4, 'Arts, Crafts & Sewing':5, 'Other': 6}
dict_category
```

```python
fig, ax = plt.subplots(1,1,figsize=(12,6))
g = sns.countplot(data = amazon_data, x = "root_bs_category", palette = "Blues", ax=ax)
for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), 
                    textcoords='offset points')
g.set_xticklabels(dict_category)
plt.xticks(rotation=45)
plt.xlabel("Ngành hàng")
plt.ylabel("Số lượng")
plt.subplots_adjust(wspace=3, hspace=3)
g.set_title("Số lượng mua hàng của mỗi ngành hàng", fontsize=20)
plt.legend(dict_category, fontsize='small')
```

```python
X_init_price = 1/amazon_data[['initial_price']] # Biến đầu vào
Y_reviews_count = amazon_data['reviews_count']
X_init_price.isna()
Y_reviews_count.isna()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_init_price,Y_reviews_count, test_size=0.2)
model = LinearRegression()
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)       

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
print(f'Intercept: {model.intercept_}')
print(f'Coefficient: {model.coef_}')
```

```python
plt.scatter(X_test, y_test, color='blue', label='Actual data')  # Dữ liệu thực tế
plt.plot(X_test, y_pred, color='red', label='Regression line')  # Đường hồi quy
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
